```python
def read_csv_line (line_nb, csv_path=''):
    
    """
    Description
    -----------
    Allows to check some lines in the csv file to spot the separator used between data.
    
    Parameters
    ----------
    line_nb: int
        Line to check in the csv file.
    csv_path: str
        Csv file's path.
        
    """
    
    opened_csv = open(csv_path)
    
    line = ""
    for line_cnt in range(line_nb+1):
        line = opened_csv.readline()
        
    print(">>>>>> Line %d" % (line_cnt))
    print(line)
```

```python
def display_EZ (x, max_rows = 100, max_cols = 100, max_colwidth = 100):
    
    """
    Description
    -----------
    Allows to display pandas dataframes with the number of rows and columns whished in an easy manner.
    
    Parameters
    ----------
    df: pandas.DataFrame()
        Dataframe to display.
    max_rows: int
        Maximum number of rows to display.
    max_cols: int
        Maximum number of columns to display.
    max_colwidth: int
        Maximum width of each column.
        
    """
    
    with pd.option_context('display.max_rows', max_rows, 'display.max_columns', max_cols, 'display.max_colwidth', max_colwidth):
        display(x)
```

```python
def count_features_dup (df):
    
    """
    Description
    -----------
    Counts the number of duplicated cells for each feature.
    
    Parameters
    ----------
    df: pandas.DataFrame()
        Dataframe to process.

    """
    
    for col in df.columns:
        n_dup = df[col].duplicated().sum()
        print("Number of duplicates in", col, "=", n_dup)
    
    return
```

```python
def count_features_empty_cells (df):
    
    """
    Description
    -----------
     Counts the number of empty cells in each feature.
    
    Parameters
    ----------
    df: pandas.DataFrame()
        Dataframe to process.

    """
    
    for col in df.columns:
        n_dup = df[col].isna().sum()
        print("Number of empty cells in", col, "=", n_dup)
    
    return
```

```python
def store_validation_values (df, validation_feature, n_sps, checked_feature = None, dups=True):
    
    """
    Description
    -----------
    Store selected indexes and features for later validation of the transformtions performed.
    
    Parameters
    ----------
    df: pandas.DataFrame()
        Dataframe to process.
    validation_feature: str
        Name of the feature chosen in which will be selected the indexes for later validation.
    n_sps: int
        Number of samples (rows) to select for validation.
    checked_feature: str
        Name of the feature to check for validation.
    dups: bool
        Select if the validation is about the management of duplicated data or something else. 
    
    Return: list
    ------
    Returns a list containing the validation values selected in the validation_feature's values.
    
    """

    if dups:
        
        # Find a duplicated row for validation.
        df_val = df.groupby(validation_feature, as_index=False).size().sort_values(by='size', ascending=False)
        validation_values = [df_val[df_val['size'] == n_sps].iloc[0,0]]
        
    else:
        
        # Select the rows for a later validation of the transformations performed on the dataframe.
        validation_values = df[df[checked_feature].isnull()].sample(n_sps)[validation_feature].to_list()
    
    # Show the validation values selected.
    print(validation_feature, "selected for validation:", str(validation_values)[1:-1])
    
    # Display the rows selected for validation.
    display(df[df[validation_feature].isin(validation_values)])

    return validation_values
```

```python
def check_dup_merge (df, df_origin, validation_feature, validation_values, dups=True): #origin_features_names = validation_feature,
    
    """
    Description
    -----------
    Checks if the transformations done performed well.
    
    Parameters
    ----------
    df: pandas.DataFrame()
        Dataframe to processed.
    df_origin: pandas.DataFrame()
        Original dataframe use ad the reference for the comparison.
    validation_feature: str
        Name of the feature chosen in which will be selected the indexes for later validation.
    validation_values: list
        Previously selected rows for comparison.
    dups: bool
        Select if the validation is about the management of duplicated data or something else. 
    
    Return: pandas.DataFrame()
    ------
    Returns the df with the comparison rows only.
    
    """

    # Check if all duplicates were removed.
    if dups:
        n_dup = df[[validation_feature]].duplicated().sum()
        print("Number of duplicated orders ID left:", n_dup)
      
    # Put the string validation value in a list for compatibility.
    if type(validation_values) != list:
        validation_values = [validation_values]
        
    # Check if the transformations happened well.
    print("\nStored", validation_feature, "for validation:\n", str(validation_values)[1:-1])

    # Display the dataset.
    print("\n\nOriginal df:")
    display(df_origin[df_origin[validation_feature].isin(validation_values)])

    # Display the dataset.
    print("\n\nProcessed df:")    
        
    return df[df[validation_feature].isin(validation_values)]
```

```python
def pie_figure (df, cat_col=None, val_col=None, figsize=(3, 3), dpi=150, explode=None):

    """
    Description
    -----------
    Plot pie chart of the chosen categories.
    
    Parameters
    ----------
    df: pandas.DataFrame()
        Dataframe to process of 1 or 2 dimensions and with categories as index or as column but with no duplicates.
    cat_col: str
        Categorical column.
    val_col: str
        Name of the column storing the values for the pie to based on.   
    figsize: tuple
        Set the size of the figure which contain the chart.
    dpi: int
        Set the resolution of the figure.
    explode:
        Set which pieces of the pie to separate from others.

    """    
    
    # If categories are already in index.
    if cat_col == None:
        
        df = df[[val_col]] # Reduce the dataframe to what is necessary.
    
    # If the pie shoudl be drawn directly on the category's column.
    elif cat_col == val_col or val_col == None:

        df = df[[cat_col]] # Reduce the dataframe to what is necessary.
        df['n_cat'] = df.count(axis=1) # Create another columns filled by 1 in front of each category label present in the row.
        df = df.groupby(cat_col).sum() # Count the number of time a category is present in the column and put the categories in index.
    
    # NB: Default case: the categories and values' columns are different and not in index.

    # Calculate the relative weight of each category.
    df = df / df.sum() * 100    
    
    # Setting of the parameters.
    plt.figure(figsize=figsize, dpi=dpi)
    labels = list(df.index)
    sizes = df.iloc[:,0]

    # Plotting
    plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%')
    plt.axis('equal')
    plt.show()
```

order_id's duplicates will be managed with the groupby() method and so it is required to no how to merge other related features together:<br>
- **payment_installments** and **payment_value** should be summed to get the total for each order ID.
- **payment_sequential:** It can be noticed that payment_sequential is different at each payment for the same order. As it is specified in the description, this feature gets an increment of 1 for each different payment type used. It is abimguous since it appears that there is an increment of 1 after each payment regardless of the payment type. I think this feature represents the number of different amount of payments done regardless of how many payments were done at the same value.<br>
**/!\ As a groupby() will be used to manage the order_id duplicates, making the sum of payments_installments will bring the same information as the highest value of payment_sequential for one order_id.<br>
=> This feature will be drop since the sum of payments_installments fits the groupby() transformation better. /!\**
- **payment_types:** Features to separate customers in distinct groups are looked for and so a categorical feature would be better if each categories represent a significant relative proportion over the others.<br>

All information to manage the order_id duplicates are now known:<br>
**- Keep features:** make a sum of payment_installments and payment_value when the corresponding order_id will be merged together and count the number of order_id duplicates as the number of payment_sequential.<br>

#### ii. Haversine distance between customers' zipcode address or customers' living states and Olist's head office

The haversine distance can be calculated between customers and the Olist's head office in order to removes negatives impact of the zipcodes categories numbering while encoding all distances in a feature alone.<br>
*NB: The haversine distance could be also calculated between sellers and customers.*

```python
from statistics import mean
```

```python
# Olist's headoffice coordinates.
olist_lat = -25.30934
olist_lng = -49.30401
```

```python
def haversine_distance (lat1, lng1, lat2, lng2, earth_r = mean([6378, 6356]), degrees=True):
    
    """
    Description
    -----------
    Calculate the haversine distance between the customers zip codes and the head office of Olist.
    
    Parameters
    ----------
    lat1: float
        Latitude of the first place (Olist's head office).
    lng1: float
        Longitude of the first place.
    lat2: float
        Latitude of the second place (Customer's zip code).
    lng2: float
        Longitude of the second place.
    earth_r: int
        Earth radius.
    degrees: bool
        Precises if the angle unit to use.
    
    Return: float
    ------
    Returns the calculated distance in the same unit as the earth radius entered.
    
    """
    
    if degrees:
        lat1, lng1, lat2, lng2 = map(np.radians, [lat1, lng1, lat2, lng2])
    
    # Haversine's formula.
    dlng = lng2 - lng1 
    dlat = lat2 - lat1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlng/2)**2
    d = 2 * earth_r * np.arcsin(np.sqrt(a))  

    return d
```

```python
%%time

# Calculation of the haversine distance between the zipcode average coordinates and the olist's head office.
df['hv_distance_zipcode_olist'] = haversine_distance(olist_lat, olist_lng, df.zipcode_lat, df.zipcode_lng)

# Calculation of the haversine distance between the cities average coordinates and the olist's head office.
df['hv_distance_city_olist'] = haversine_distance(olist_lat, olist_lng, df.city_lat, df.city_lng)

# Calculation of the haversine distance between the states average coordinates and the olist's head office.
df['hv_distance_state_olist'] = haversine_distance(olist_lat, olist_lng, df.state_lat, df.state_lng)
```
